```python
# Initialize values
target_name = "DUAT"   # replace with your SURNAME in uppercase
SEED_NUM = 4           # replace with last digit of your student ID
redacted_name = ""

# Process the surname
for char in target_name:
    if char in "AEIOU":
        redacted_name += str(SEED_NUM)
    else:
        redacted_name += char

# Output
print("Original Surname:", target_name)
print("SEED_NUM Used:", SEED_NUM)
print("Redacted Output:", redacted_name)
```

```python
# Engineer name
engineer_name = "JHON"   # your first name

# Get first letter
first_letter = engineer_name[0]

# Convert to ASCII
ascii_value = ord(first_letter)

# Total loops from Task 1
loops = len(target_name)

# Calculate hash
verification_hash = (ascii_value * SEED_NUM) - loops

# Output
print("First Letter ASCII (A):", ascii_value)
print("SEED_NUM (B):", SEED_NUM)
print("Total Loops (C):", loops)
print("Verification Hash:", verification_hash)
```
